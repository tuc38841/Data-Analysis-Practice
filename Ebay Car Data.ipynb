```python
import numpy as np
import pandas as pd
```

```python
autos = pd.read_csv('autos.csv', encoding='Latin-1')
```

```python
# Update col names to snakecase and succinct
autos.rename({'dateCrawled':'date_crawled','offerType':'offer_type','vehicleType':'vehicle_type',
              'yearOfRegistration':'registration_year','powerPS':'power_ps','monthOfRegistration':'registration_month',
              'fuelType':'fuel_type','notRepairedDamage':'unrepaired_damage','dateCreated':'ad_created',
              'nrOfPictures':'num_of_pictures','postalCode':'postal_code','lastSeen':'last_seen'},
             axis=1, inplace=True)
```

```python
# Examine stats on columns to look for lack of structure
autos.describe(include='all')
autos.info()
#cols to convert to int/float = price, odometer
#num_of_pics = no values (all zero)
#seller and offer_type all the same
#unreparied_damage = contains german('nein')
```

```python
#Change price and odometer values to numeric
autos.rename({'price':'price_dollar','odometer':'odometer_km'},
             axis=1,inplace=True)
autos['price_dollar'] = (autos['price_dollar'].str.replace('$','')
                                            .str.replace(',','')
                                            .str.replace("'","")
                                            .astype(int)
                        )
autos['odometer_km'] = (autos['odometer_km'].str.replace('km','')
                                            .str.replace(',','')
                                            .str.replace("'","")
                                            .astype(int)
                       )
```

```python
# Remove cols with no values or little relevance
autos.drop(['seller','offer_type','num_of_pictures'],axis=1)
```

```python
autos['registration_year'].value_counts(normalize=True)
old_registration = autos['registration_year']

for reg in old_registration:
    if (reg < 1900) | (reg > 2016):
        questionable_reg_indexes = autos[autos['registration_year'] == reg].index
        autos.drop(questionable_reg_indexes,inplace=True)
    else:
        pass
# Some values in registration_year were greater than 2016 or 
# less than 1900. Found and removed those entries from dataframe
```

```python
# Aggregate based on brand of vehicle
unique_brands = autos['brand'].unique()
unique_brands
```

```python
# Get average price for top 20 brand vehicles
top_20_brands = autos['brand'].value_counts().head(20).keys()
```

```python
avg_price_per_brand = {}

for brand in top_20_brands:
    avg_price = round((autos[autos['brand'] == brand]['price_dollar'].sum()) / (len(autos[autos['brand'] == brand])))
    avg_price_per_brand[brand] = int(avg_price)
    
avg_price_per_brand
```

```python
# Average milage per top 20 brand vehicles
avg_milage_per_brand = {}

for brand in top_20_brands:
    milage = 0.621 * (autos[autos['brand'] == brand]['odometer_km'].sum())
    mean_milage = round(milage / len(autos[autos['brand'] == brand]))
    avg_milage_per_brand[brand] = int(mean_milage)
    
avg_milage_per_brand
```

```python
# Combine price and milage per top 20 brands
# by creating 2 series and then dataframe of both

price_series = pd.Series(avg_price_per_brand)
milage_series = pd.Series(avg_milage_per_brand)

price_milage_df = pd.DataFrame(price_series, columns=['mean_price'])
price_milage_df['mean_milage'] = milage_series
price_milage_df
```
